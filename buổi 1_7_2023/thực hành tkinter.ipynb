```python
from tkinter import *
from tkinter import messagebox

window = Tk()
window.title("book management")

book_list = []

def add_book():
    name_volue = name.get()
    category_volue = category.get()
    price_volue = price.get()
    if name_volue == '' or category_volue == '' or price_volue == '':
        messagebox.showerror("Add Book","please enter name, category and price!")
    else:
        new_book = Book(name_volue,category_volue,int(price_volue))
        book_list.append(new_book)
        print([book.name for book in book_list])
        messagebox.showinfo("Add Book",f"Add {name_volue} successfully")
        
def search_book():
    search_name = search.get()
    check = False
    
    for i in range(len(book_list)):
        if book_list[i].name == search_name:
            messagebox.showinfo("Search Book", book_list[i])
            check = True
            break
            
    if check == False:
        messagebox.showerror("Search Book", f"Cannot find '{search_name}'!")
        
def delete_book():
    delete_name = search.get()
    check = False
    
    for i in range(len(book_list)):
        if book_list[i].name == delete_name:
            del book_list[i]
            print([book.name for book in book_list])
            messagebox.showinfo("Delete Book", f"Delete {delete_name} successfully!")
            check = True
            break
            
    if check == False:
        messagebox.showerror("Delete Book", f"Cannot find '{delete_name}'!")
        
def update_price_book():
    update_price_name = search.get()
    new_price = update_price.get()
    check = False
    for i in range(len(book_list)):
        if book_list[i].name == update_price_name:
            book_list[i].price = new_price
            print([book.name for book in book_list])
            messagebox.showinfo("Update Book",book_list[i])
            check = True
            break
            
    if check == False:
        messagebox.showerror("Update Book",f"canot find '{update_price_name}'!")
    
lbl_name = Label(window,text = "book name: ")
lbl_category = Label(window,text = "book category: ")
lbl_price = Label(window,text="book price: ")

name = Entry(window)
category = Entry(window)
price = Entry(window)
search = Entry(window)
update_price = Entry(window)

btn_add = Button(window,text = "Add", command = add_book)
btn_search = Button(window,text = "Search", command = search_book)
btn_delete = Button(window,text = "Delete", command = delete_book)
btn_update_price = Button(window,text = "Update Price", command = update_price_book)

lbl_name.grid(row = 0,column = 0)
name.grid(row = 0,column = 1)
lbl_category.grid(row = 1,column = 0)
category.grid(row = 1,column = 1)
lbl_price.grid(row = 2,column = 0)
price.grid(row = 2,column = 1)
btn_add.grid(row = 3,column = 0)
search.grid(row = 4,column = 0)
btn_search.grid(row = 4,column = 1)
btn_delete.grid(row = 4,column = 2)
update_price.grid(row = 5,column = 0)
btn_update_price.grid(row = 5,column = 1)


window.mainloop()
```

```python
class Book:
    def __init__(self, name, category,price):
        self.name = name
        self.category=category
        self.price=price
        
    def __str__(self):
        return f"Name: {self.name}, category: {self.category}, price: {self.price}"
```
