### Bài 6: Học viên sẽ không đủ điều kiện thi nếu tỉ lệ có mặt nhỏ hơn 75%. Viết chương trình nhập vào tổng số buổi học của một khóa học và tổng số buổi học mà học viên tham gia. Sau đó cho biết học viên đó có đủ điều kiện thi hay không.

````python
# a = Tổng buổi học trong một khóa
# b = Tổng số buổi học mà học viên tham gia
a = int(input("Số buổi học trong một khóa: "))
b = int(input("Số buổi học mà học viên tham gia: "))
if (b / a) >= 0.75: print("học viên đủ điều kiện tham gia thi")
else: print("Học viên không đủ điều kiện tham gia thi")
````

### Bài 11:

Nhập vào chiều cao (cm) và cân nặng (kg), tính số BMI và xét rồi xin kết quả theo dữ liệu sau:

- BMI < 16: Gầy cấp độ III
- 16 <= BMI < 17:  Gầy cấp độ II
- 17<= BMI < 18.5: Gầy cấp độ I
- 18.5 <= BMI < 25: Bình thường
- 25 <= BMI < 30: Thừa cân
- 30 <= BMI < 35 : Béo phì cấp độ I
- 35 <= BMI < 40: Béo phì cấp độ II
- BMI > 40: Béo phì cấp độ III

BMI = Cân nặng / (chiều cao) * 2,  chiều cao(m)

````python
a = int(input("nhập chiều cao: "))
b = int(input("nhập cân nặng: "))
c = b / ((a / 100) ** 2) 
if c < 16: print("gay cap do 3")
elif c >= 16 and c < 17: print("gay cap do 2")
elif c >= 17 and c < 18.5: print("gay cap do 1")
elif c >= 18.5 and c < 25: print("binh thuong")
elif c >= 25 and c < 30: print("thua can")
elif c >= 30 and c < 35: print("beo phi cap do 1")
elif c >= 35 and c < 40: print("beo phi cap do 2")
else: print("beo phi cap do 3")
print(f"{c}")
````
