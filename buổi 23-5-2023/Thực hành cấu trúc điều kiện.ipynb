### Bài 10:
Tại một cửa hàng nọ có menu như sau:

3. Pesi giá 10k

Tính và in ra tổng giá tiền

Nếu giá trị đơn hàng trên 2 triệu, giảm 10% cho khách, trên 4 triệu, giảm 15%

Tính và in ra tổng giá tiền được giảm

Tính và in ra tổng giá tiền khách cần trả

In ra dòng chữ "Cảm ơn quý khách đã mua hàng"

gà rán = 35
hamburger = 50
pesi = 10

````python
"""Ví dụ:
Nhập số lượng gà rán khách mua:100
Nhập số lượng hamburger khách mua:100
Nhập số lượng pesi khách mua:15
Tổng giá tiền đơn hàng là: 8.650.000
Tổng tiền được giảm là: 1.297.500
Tổng tiền phải thanh toán: 7.352.500
Cảm ơn quý khách đã mua
"""
````

````python
garan=int(input("nhập số gà rán khách mua: "))
hamburger=int(input("nhập số hamburger khách mua:"))
pesi=int(input("nhập số pesi khách mua: "))
tong=(garan*35000)+(hamburger*50000)+(pesi*10000)
print("tổng số tiền đơn hàng là:",tong)
if 4000000>tong>2000000:
    giam=tong*10/100
    print("tổng tiền được giảm là",giam)
    print("tổng tiền phải thanh toán là",tong-giam)
elif tong>=4000000:
    giam=tong*15/100
    print("tổng tiền được giảm là",giam)
    print("tổng tiền phải thanh toán là",tong-giam)
print("cảm ơn quý khách dã mua")
````

### Bài 11:

Nhập vào chiều cao (cm) và cân nặng (kg), tính số BMI và xét rồi xin kết quả theo dữ liệu sau:

- BMI < 16: Gầy cấp độ III
- 16 <= BMI < 17:  Gầy cấp độ II
- 17<= BMI < 18.5: Gầy cấp độ I
- 18.5 <= BMI < 25: Bình thường
- 25 <= BMI < 30: Thừa cân
- 30 <= BMI < 35 : Béo phì cấp độ I
- 35 <= BMI < 40: Béo phì cấp độ II
- BMI > 40: Béo phì cấp độ III

BMI = Cân nặng / (chiều cao) * 2,  chiều cao(m)

````python
cao = int(input("nhập chiều cao: "))
nang = int(input("nhập cân nặng: "))
BMI = nang / ((cao / 100)*2) 
if BMI < 16: print("gay cap do 3")
elif 16 <= BMI < 17: print("gay cap do 2")
elif 17<= BMI < 18.5: print("gay cap do 1")
elif 18.5 <= BMI < 25: print("binh thuong")
elif 25 <= BMI < 30: print("thua can")
elif 30 <= BMI < 35: print("beo phi cap do 1")
elif 35 <= BMI < 40: print("beo phi cap do 2")
else: print("beo phi cap do 3")
print(f"{BMI}")
````
