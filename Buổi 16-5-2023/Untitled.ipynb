```python
t=int(input("nhập điểm toán: "))
l=int(input("nhập điểm lý: "))
h=int(input("nhập điểm hóa: "))
DTB=(t+l+h)/3
print(DTB)
if DTB>8:
    print("điểm trung bình nằm trong mức giỏi")
if 6<=DTB<8:
    print("điểm trung bình nằm trong mức khá")
if DTB<6:
    print("điểm trung bình nằm không nằm trong mức trung bình")
```
