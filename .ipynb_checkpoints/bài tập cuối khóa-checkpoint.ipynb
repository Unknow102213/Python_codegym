```python
import tkinter as tk
from tkinter import *
from tkinter import messagebox,ttk

window = tk.Tk()
window.title("personal information management")
window.geometry('1230x500')

#treeview
columns = ('name','age','sex','address','sdt','email')
tree = ttk.Treeview(window, columns = columns, show = 'headings')
tree.heading('name', text = 'NAME')
tree.heading('age', text = 'AGE')
tree.heading('sex', text = 'SEX')
tree.heading('address', text = 'ADDRESS')
tree.heading('sdt', text = 'SĐT')
tree.heading('email', text = 'EMAIL')

#Add Information
def add_information():
    name_value = name.get()
    age_value = age.get()
    sex_value = sex.get()
    address_value = address.get()
    sdt_value = sdt.get()
    email_value = email.get()
    if name_value == '' or age_value == '' or sex_value == '' or address_value == '' or sdt_value == '' or email_value == '':
        messagebox.showerror("Add Information","please enter all information!")
    else:
        tree.insert('',0,values =(name_value,age_value,sex_value,address_value,sdt_value,email_value))
        messagebox.showinfo("Add Information", f"Add {name_value} successfully!")

#Search Information
def search_information():
    search_name = search.get()
    check = False
    for row_id in tree.get_children():
        value = tree.item(row_id)["values"]
        if value[0] == search_name:
            messagebox.showinfo("Search information",f"Name:{value[0]},Age: {value[1]},Sex: {value[2]},Address: {value[3]},Sđt: {value[4]},Email: {value[5]}")
            check = True
            break

    if check == False:
        messagebox.showerror("Search information", f"Cannot find '{search_name}'!")

#Delete Information
def delete_information():
    check = False
    for select_item in tree.selection():
        tree.delete(select_item)
        messagebox.showinfo("Delete Information", f"Delete successfully!")
        check = True
    if check == False:
        messagebox.showerror("Delete Information","please choose Information in the board to delete!")

#Update Information
def update_information():
    new_name = update_name.get()
    new_age = update_age.get()
    new_sex = update_sex.get()
    new_address = update_address.get()
    new_sdt = update_sdt.get()
    new_email = update_email.get()
    check = False

    for select_item in tree.selection():
            item = tree.item(select_item)
            record = item['values']
            if new_name == '' or new_name == record[0]:
                record[0] = record[0]
            elif record[0] != new_name:
                record[0] = new_name
            if new_age == '' or new_age == record[1]:
                record[1]= record[1]
            elif record[1] != new_age:
                record[1] = new_age
            if new_sex == '' or new_sex == record[2]:
                record[2] = record[2]
            elif sex != new_sex:
                record[2] = new_sex
            if new_address == '' or new_address == record[3]:
                record[3] = record[3]
            elif record[3] != new_address:
                record[3] = new_address
            if new_sdt == '' or new_sdt == record[4]:
                record[4] = record[4]
            elif record[4] != new_sdt:
                record[4] = new_sdt
            if new_email == '' or new_email == record[5]:
                record[5] = record[5]
            elif record[5] != new_email:
                record[5] = new_email
            tree.delete(select_item)
            tree.insert('',tk.END,values=(record[0],record[1],record[2],record[3],record[4],record[5]))
            messagebox.showinfo("Update Information","update successfully!")
            check=True
    if check == False:
        messagebox.showerror("Update Information","Please choose Information in the board to update")

# Scrollbar
scrollbar = ttk.Scrollbar(window, orient = tk.VERTICAL, command = tree.yview)
tree.configure(yscroll = scrollbar.set)
scrollbar.grid(row = 0, column = 1,sticky = 'ns')

#Label
lbl_add = Label(window, text = "ADD PERSONAL INFORMATION",fg ='green')
lbl_attention_add = Label(window, text = "Note: Please enter all information to add!",fg ='red')
lbl_attention_add.config(font=("Courier", 10))
lbl_name = Label(window, text = "Name: ")
lbl_age = Label(window,text = "Age: ")
lbl_sex = Label(window,text = "Sex: ")
lbl_address = Label(window,text = "Address: ")
lbl_sdt = Label(window,text = "Sđt: ")
lbl_email = Label(window,text = "Email")
lbl_search=Label(window,text = "SEARCH PERSONAL INFORMATION",fg = 'green')
lbl_attention_search=Label(window,text = "Note: please enter name to search!",fg='red')
lbl_attention_search.config(font=("Courier",10))
lbl_delete = Label(window, text = "DELETE PERSONAL INFORMATION",fg = 'green')
lbl_attention_delete = Label(window, text = "please choose Information in the board to delete!",fg = 'red')
lbl_attention_delete.config(font=("Courier", 10))
lbl_update = Label(window, text = "UPDATE PERSONAL INFORMATION",fg = 'green')
lbl_attention_update = Label(window, text = "please choose Information in the board to update!",fg = 'red')
lbl_attention_update.config(font=("Courier", 10))
lbl_name_to_update = Label(window, text = "Name to update:")
lbl_update_name = Label(window, text = "New Name:")
lbl_update_age = Label(window, text = "New Age:")
lbl_update_sex = Label(window, text = "New Sex:")
lbl_update_address = Label(window, text = "New Address:")
lbl_update_sdt = Label(window, text = "New Sđt:")
lbl_update_email = Label(window, text = "New Email:")


#Entry
name = Entry(window)
age = Entry(window)
sex = Entry(window)
address = Entry(window)
sdt = Entry(window)
email = Entry(window)
search = Entry(window)
update_name = Entry(window)
update_age = Entry(window)
update_sex = Entry(window)
update_address = Entry(window)
update_sdt = Entry(window)
update_email = Entry(window)

#Button
btn_add = Button(window, text = "Add",command = add_information)
btn_search = Button(window, text = "Search",command = search_information)
btn_delete = Button(window, text = "Delete",command = delete_information)
btn_update = Button(window, text = "Update",command = update_information)

#đặt vị trí
tree.grid(row = 0, column = 0, sticky = 'nsew')
lbl_add.place(x = 30, y = 250)
lbl_attention_add.place(x = 0, y = 270)
lbl_name.place(x = 10, y = 300)
name.place(x = 100, y = 300)
lbl_age.place(x = 10, y = 320)
age.place(x = 100, y = 320)
lbl_sex.place(x = 10, y = 340)
sex.place(x = 100, y = 340)
lbl_address.place(x = 10, y = 360)
address.place(x = 100, y = 360)
lbl_sdt.place(x = 10, y = 380)
sdt.place(x = 100, y = 380)
lbl_email.place(x = 10, y = 400)
email.place(x = 100, y = 400)
btn_add.place(x = 100, y = 450)
lbl_search.place(x = 380, y = 250)
lbl_attention_search.place(x=380,y=270)
search.place(x = 380, y = 320)
btn_search.place(x = 530, y = 320)
lbl_delete.place(x = 380, y = 380)
lbl_attention_delete.place(x=380,y=400)
btn_delete.place(x = 530, y = 450)
lbl_update.place(x = 800, y = 250)
lbl_attention_update.place(x=800,y=270)
lbl_update_name.place(x=800,y=300)
update_name.place(x = 900, y = 300)
lbl_update_age.place(x = 800, y = 320)
update_age.place(x = 900,y = 320)
lbl_update_sex.place(x = 800,y = 340)
update_sex.place(x = 900,y = 340)
lbl_update_address.place(x = 800,y = 360)
update_address.place(x = 900,y = 360)
lbl_update_sdt.place(x = 800,y = 380)
update_sdt.place(x = 900,y = 380)
lbl_update_email.place(x = 800,y = 400)
update_email.place(x = 900,y = 400)
btn_update.place(x = 900,y = 450)

#chạy window
window.mainloop()
```
